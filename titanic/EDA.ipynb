```python
# 'Age', 'Pclass', 'Survived'
sns.displot(x = 'Age', col = 'Pclass', hue = 'Survived', data = train, multiple = 'stack', height = 5, aspect = 1)
```

- **Pclass**는 Ticket Class를 나타내는데, 연령대가 높으면 Ticket Class가 높아지고, 연령대가 낮으면 Ticket Class 또한 낮아짐을 볼 수 있다.


- **Pclass**와 **Alone**에 따라서 **Age**의 결측값을 채워주도록 하자.

```python
print('*Before* Amount of Null Data *Train Age* :', train['Age'].isnull().sum(), '\n')
print('*Before* Amount of Null Data *Test Age* :', test['Age'].isnull().sum(), '\n')

for alone in train['Alone'].unique():
    for pclass in [1, 2, 3]:
        
        # 'Age' 결측치를 제거한 데이터 추출
        data = train.dropna(subset = ['Age'])
        
        # 'Alone', 'Pclass'에 따른 'Age'의 중앙값 추출
        value = data[(data['Alone'] == alone) & (data['Pclass'] == pclass)]['Age'].median()
        
        # 추출한 값을 Train과 Test 데이터 'Age' 결측값에 대입
        train.loc[(train['Age'].isnull()) & (train['Alone'] == alone) & (train['Pclass'] == pclass), 'Age'] = value
        test.loc[(test['Age'].isnull()) & (test['Alone'] == alone) & (test['Pclass'] == pclass), 'Age'] = value
        
print('*After* Amount of Null Data *Train Age* :', train['Age'].isnull().sum(), '\n')
print('*After* Amount of Null Data *Test Age* :', test['Age'].isnull().sum())
```

```python
# Age 변수를 5개로 구간화
train['Band_age'] = pd.cut(train['Age'], 5)

# 시각화를 통해 살펴보기
plt.figure(figsize = (16, 6))

plot_data = pd.pivot_table(train, index = 'Band_age', values = 'Survived', aggfunc = 'mean').sort_index()
sns.barplot(x = plot_data.index, y = 'Survived', data = plot_data)
plt.title('Survival Rate By Band_age')
plt.xlabel('')
plt.ylabel('')
```
