```python
import pandas as pd
import warnings
warnings.filterwarnings(action = 'ignore')
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('타이타닉.train.csv')
test = pd.read_csv('타이타닉.test.csv')
```

```python
'''
해야할 것들
1. 분류모델 평가지표 공부
2. 더미화 공부
'''
```

```python
target=train['Survived']
train=train.drop(['Name','PassengerId','Fare','SibSp','Parch','Survived','Family','Embarked'], axis=1)
```

```python
train.head(3)
```

```python
dummies = ['Title','Pclass','Embarked_binning','Fare_binning']
for i in range(len(dummies)):
    train_mat=pd.get_dummies(train.loc[:,dummies[i]], drop_first = True, prefix = dummies[i])
    train = train.join(train_mat)
```

```python
train
```

```python
train=train.drop(['Title','Pclass','Embarked_binning','Fare_binning'], axis=1)
```

```python
train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from xgboost import plot_importance
from xgboost import XGBClassifier
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

```python
model = RandomForestClassifier()
model.fit(train,target)
scoring = 'accuracy'
score = cross_val_score(model, train, target, cv = k_fold, scoring = scoring)
print(score)
print(score.mean())
train_pred=model.predict(train)
```

```python
model = XGBClassifier()
model.fit(train,target)
scoring = 'accuracy'
score = cross_val_score(model, train, target, cv = k_fold, scoring = scoring)
print(score)
print(score.mean())
train_pred=model.predict(train)
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(target,train_pred)
sns.heatmap(mat, square=True, annot=True, cbar = False ,fmt = '' )

plt.xlabel('predicted value')
plt.ylabel('true values')
```

- 생존자를 잘 예측하지 못하는 것은 아닐까?
- F1 score을 이용하여 검증값을 구해보자.
- accuracy를 이용하여 검증값을 구해보자.

```python
from yellowbrick.classifier import ROCAUC
visualizer = ROCAUC(model, micro=False, macro = False)
visualizer.fit(train, target)
visualizer.score(train, target)
visualizer.show()
```

- 그래프에서 ROCAUC를 확인해보면 재현율과 특이도가 1에 가까운 것을 확인할 수 있다.

- x축은 False Positive Rate
- y축은 True Positive Rate

- 지도학습의 경우 ROCAUC그래프를 사용하여 구할 수 있다.
- 비지도 학습의 경우 KElbowVisualier를 통해서 구할 수 있다.

- 임계값을 낮게 설정하는 경우 우리는 False Positive 가 늘어난다.
- 임계값을 높게 설정하는 경우 우리는 False negative 가 늘어난다.

- Roc curve의 휜 정도는 우리가 ex) 암에 걸린사람과 안걸린 사람을 잘 구별했다는 것을 알 수 있다.
- 그래서 Roc curve의 면적이 클수록 = Auc가 클수록 잘 구별했다고 생각하면 된다.

- 1. 데이터가 전부 비슷한 상태라면 accuracy를 사용해도 상관이 없다.
- 2. 그러나 데이터의 수가 차이가 많이 난다면 accuracy에서 잘못된 분석이 될 수 있다.
- 3. 이러할 때는 조화평균을 이용한 F1 score을 사용하여 검증을 하도록 하자.

-이유:
데이터의 수가 차이가 많이 난다면 많은 변수에서 accuracy가 높으면 accuracy 값이 높다고 나올 확률이 있다.
그러나 f1 score같은경우 조화평균을 사용하는데 조화평균은 적은 데이터를 기준으로 평균을 구하기 때문에 전체적으로 모델링이 잘되어있는지확인할 수 있다.
즉 데이터의 수가 차이가 많이 난다면 F1 score을 이용해야 한다.
데이터의 수가 차이가 없다면 accuracy를 이용하여 사용해도 상관이 없다.
