# Description
- What sorts of people were more likely to survive?


- Use machine learning to create a model that predicts which passengers survived the Titanic shipwreck.

# Data Fields
1. **Survived** - Survival	
    - 0 = No
    - 1 = Yes
    
    
2. **Pclass** - Ticket class, A proxy for socio-economic status
    - 1 = 1st, Upper
    - 2 = 2nd, Middle
    - 3 = 3rd, Lower


3. **Sex** - Sex	


4. **Age** - Age in years (Age is fractional if less than 1)


5. **Sibsp** - 함께 탑승한 형제와 배우자의 수	


6. **Parch** - 함께 탑승한 부모, 아이의 수	


7. **Ticket** - Ticket number	


8. **Fare** - Passenger fare	


9. **Cabin** - Cabin number	


10. **Embarked** - Port of Embarkation(탑승 항구)
    - C = Cherbourg
    - Q = Queenstown
    - S = Southampton

## [0]. 라이브러리 불러오기

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import missingno as msno

%matplotlib inline
plt.style.use('ggplot')
plt.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False

import warnings
warnings.filterwarnings(action = 'ignore')
```

## [1]. 데이터 적재

```python
train = pd.read_csv('train.csv')
print('Train Data의 행과 열의 개수 :', train.shape, '\n')

test = pd.read_csv('test.csv')
print('Test Data의 행과 열의 개수 :', test.shape)

# train과 test를 리스트로 묶어줌으로써 나중에 변수를 수정하거나 삭제할 때 유용하다.
combine = [train, test]
```

## [2]. 데이터 탐색

```python
# Train Data 일부 확인하기
train.head(3)
```

```python
# Test Data 일부 확인하기
test.head(3)
```

```python
print(train.info(), '\n')
print('-' * 50)
print(test.info())
```

```python
# Train, Test Data 결측치 시각화로 확인하기
fig, axes = plt.subplots(ncols = 2, figsize = (16, 7))

msno.bar(train, color = (0.8, 0.5, 0.2), fontsize = 13, ax = axes[0])
axes[0].set_title('Null Data Of Train')

msno.bar(test, color = (0.5, 0.8, 0.2), fontsize = 13, ax = axes[1])
axes[1].set_title('Null Data Of Test')

fig.tight_layout()
```

**[Train]**

- **Train** 데이터의 경우 **Age, Cabin** 두 개의 변수에 결측치가 있음을 볼 수 있다.


- **Cabin** 변수는 대부분의 관측치가 결측치 임을 볼 수 있다.


- 따라서, **Cabin** 변수를 사용하는 것은 적절하지 않은것으로 판단된다.

**[Test]**

- **Test** 데이터의 경우 **Age, Fare, Cabin** 세 개의 변수에 결측치가 있음을 볼 수 있다.


- 마찬가지로, Test 데이터에도 **Cabin** 변수는 대부분 결측치 임을 볼 수 있다.


- **PassengerId, Cabin** 두 개의 변수는 삭제해주도록 하자.

```python
# PassengerId, Cabin 두 개의 변수 삭제
print('Shape Of Train Data *Before Delete* :', train.shape, '\n')
print('Shape Of Test Data *Before Delete* :', test.shape, '\n')

train.drop(['PassengerId', 'Cabin'], axis = 1, inplace = True)
test.drop(['PassengerId', 'Cabin'], axis = 1, inplace = True)
combine = [train, test]

print('Shape Of Train Data *After Delete* :', train.shape, '\n')
print('Shape Of Test Data *After Delete* :', test.shape)
```

**Sex**

```python
fig, axes = plt.subplots(ncols = 2, figsize = (16, 6))

plot_data1 = train['Sex'].value_counts()
plot_data1.plot.pie(explode = [0, 0.03], shadow = True, autopct = '%.1f%%', fontsize = 15, ax = axes[0])
axes[0].set_ylabel('')
axes[0].set_title('Pie Chart Of Sex')

plot_data2 = pd.pivot_table(train, index = 'Sex', values = 'Survived', aggfunc = 'mean')
sns.barplot(x = plot_data2.index, y = 'Survived', data = plot_data2, ax = axes[1])
axes[1].set_title('Survived Rate By Sex')
axes[1].set_ylabel('')
axes[1].set_xlabel('')
```

- 여성 탑승객에 비해 남성 탑승객이 대략 **2배** 많음을 볼 수 있다.


- 반면에, 남성의 생존률은 여성에 비해 매우 낮음을 볼 수 있다.


- 아마도, 여성을 먼저 구출하지 않았을까 라는 추측을 할 수 있다.

**Parch**

- 함께 탑승한 부모, 아이의 수

**SibSp**

- 함께 탑승한 형제와 배우자의 수

```python
# 'Parch', 'SibSp'의 분포
fig, axes = plt.subplots(nrows = 2, figsize = (16, 9))

plot_data1 = pd.DataFrame(train['Parch'].value_counts()).sort_index()
plot_data2 = pd.DataFrame(train['SibSp'].value_counts()).sort_index()

sns.barplot(x = plot_data1.index, y = 'Parch', data = plot_data1, ax = axes[0])
axes[0].set_title('Barplot Of Parch')
axes[0].set_ylabel('')

for index, [value] in enumerate(plot_data1.values):
    axes[0].text(index, value, value, ha = 'center', fontsize = 15)

sns.barplot(x = plot_data2.index, y = 'SibSp', data = plot_data2, ax = axes[1])
axes[1].set_title('Barplot Of SibSp')
axes[1].set_ylabel('')

for index, [value] in enumerate(plot_data2.values):
    axes[1].text(index, value, value, ha = 'center', fontsize = 15)
    
fig.tight_layout()
```

- 위의 두 그림을 보고, 대부분의 탑승객이 혼자 탑승했음을 볼 수 있다.


- 함께 탑승한 가족 수에 따라, 생존률이 변하는지 확인해보도록 하자.

```python
# 생존률 확인
fig, axes = plt.subplots(ncols = 2, figsize = (16, 6))

plot_data1 = pd.pivot_table(train, index = 'Parch', values = 'Survived', aggfunc = 'mean')
plot_data2 = pd.pivot_table(train, index = 'SibSp', values = 'Survived', aggfunc = 'mean')

sns.barplot(x = plot_data1.index, y = 'Survived', data = plot_data1, ax = axes[0])
axes[0].set_title('Survived Rate By Parch')
axes[0].set_ylabel('')
axes[0].set_xlabel('')

sns.barplot(x = plot_data2.index, y = 'Survived', data = plot_data2, ax = axes[1])
axes[1].set_title('Survived Rate By SibSp')
axes[1].set_ylabel('')
axes[1].set_xlabel('')

fig.tight_layout()
```

- **Parch, SibSp**의 수에 따라서 특정한 패턴이 보이지 않는다.


- 따라서, **Family = Parch + SibSp + 1** 가족의 수를 의미하는 변수를 만들어주자.

```python
# Family 변수 생성
print('Shape Of Train *Before Create* :', train.shape, '\n')
print('Shape Of Test *Before Create* :', test.shape, '\n')

for dataset in combine:
    dataset['Family'] = dataset['Parch'] + dataset['SibSp'] + 1
    
print('Shpae Of Train *After Create* :', train.shape, '\n')
print('Shape Of Test *After Create* :', test.shape)
```

```python
# Family의 수에 따른 생존률 확인
plt.figure(figsize = (16, 5))

plot_data = pd.pivot_table(train, index = 'Family', values = 'Survived', aggfunc = 'mean')

sns.barplot(x = plot_data.index, y = 'Survived', data = plot_data)
plt.xlabel('')
plt.ylabel('')
plt.title('Survived Rate By Family')
```

- **Family** 변수를 만들어줬음에도 불구하고, Family 수에 따른 생존률에 특정한 패턴을 찾을 수 없다.


- 따라서, 위의 그림에서 혼자 탑승한 승객이 많았으므로 혼자 탑승한 승객과 그렇지 않은 승객을 구분해주는 변수를 만들어보자.

```python
# 'Alone' 변수 만들기
print('Shape Of Train *Before Create* :', train.shape, '\n')
print('Shape Of Test *Before Create* :', test.shape, '\n')

for dataset in combine:
    dataset['Alone'] = 0
    dataset.loc[dataset['Family'] == 1, 'Alone'] = 1
    
print('Shpae Of Train *After Create* :', train.shape, '\n')
print('Shape Of Test *After Create* :', test.shape)
```

```python
# Alone의 분포와 Alone에 따른 생존률 확인
fig, axes = plt.subplots(ncols = 2, figsize = (16, 6))

plot_data1 = train['Alone'].value_counts().sort_index()
plot_data1.plot.pie(explode = [0, 0.05], shadow = True, autopct = '%.1f%%', 
                    fontsize = 15, labels = ['Not Alone', 'Alone'], ax = axes[0])
axes[0].set_ylabel('')
axes[0].set_title('Pie Chart Of Alone')

plot_data2 = pd.pivot_table(train, index = 'Alone', values = 'Survived', aggfunc = 'mean')
sns.barplot(x = plot_data2.index, y = 'Survived', data = plot_data2, ax = axes[1])
axes[1].set_xlabel('')
axes[1].set_ylabel('')
axes[1].set_xticklabels(['Not Alone', 'Alone'])
axes[1].set_title('Survived Rate By Alone')
```

**Age**

```python
# 'Age'의 분포 확인
fig, axes = plt.subplots(nrows = 2, figsize = (16, 8))

sns.distplot(train['Age'], color = 'black', ax = axes[0])
axes[0].set_xlabel('')
axes[0].set_ylabel('')
axes[0].set_title('Distplot Of Age')

sns.distplot(train[train['Survived'] == 0]['Age'], hist = False, color = 'red', label = 'No', ax = axes[1])
sns.distplot(train[train['Survived'] == 1]['Age'], hist = False, color = 'blue', label = 'Yes', ax = axes[1])
axes[1].set_xlabel('')
axes[1].set_ylabel('')
axes[1].set_title('Distplot Of Age By Survived')
axes[1].legend()

fig.tight_layout()
```

- 전체 탑승객의 나이 분포를 살펴보면 주로 **20 ~ 40**대 임을 볼 수 있다.

**Name**

```python
# Name 변수를 사용하여 새로운 변수 만들기
print('Shape Of Train *Before Create* :', train.shape, '\n')
print('Shape Of Test *Before Create* :', test.shape, '\n')

for dataset in combine:
    dataset['Title'] = dataset['Name'].str.extract('([A-Za-z]+)\.', expand = False)

print('Shape Of Train *After Create* :', train.shape, '\n')
print('Shape Of Test *After Create* :', test.shape)
```

```python
# Train Data : 'Sex'와 'Title' 교차표 확인
pd.crosstab(train['Sex'], train['Title']).style.background_gradient(cmap = 'Blues')
```

```python
# Test Data : 'Sex'와 'Title' 교차표 확인
pd.crosstab(test['Sex'], test['Title']).style.background_gradient(cmap = 'Greens')
```

- 영어의 호칭에는 **Mr, Mrs, Miss, Ms** 등이 있다.


- 위의 교차표를 보고 **Rare, Mr, Mrs, Miss** 네 개의 범주로 줄여주도록 하자.

```python
print('Length Of Unique Train Title *Before Modify* :', len(train['Title'].unique()), '\n')
print('Length Of Unique Test Title *Before Modify* :', len(test['Title'].unique()), '\n')

for dataset in combine:
    dataset['Title'] = dataset['Title'].replace(['Countess', 'Capt', 'Col', 'Don', 'Dr', 'Jonkheer', 'Major',
                                                'Rev', 'Sir', 'Dona'], 'Rare')
    dataset['Title'] = dataset['Title'].replace('Master', 'Mr')
    dataset['Title'] = dataset['Title'].replace(['Lady',  'Mlle', 'Mme', 'Ms'], 'Mrs')
    
print('Length Of Unique Train Title *After Modify* :', len(train['Title'].unique()), '\n')
print('Length Of Unique Test Title *After Modify* :', len(test['Title'].unique()))
```

```python
# 'Title'의 분포와 'Title'에 따른 생존률 시각화
fig, axes = plt.subplots(ncols = 2, figsize = (16, 6))

plot_data = pd.DataFrame(train['Title'].value_counts()).sort_index()
plot_data.plot.pie(explode = [0, 0.02, 0.02, 0.02], shadow = True, autopct = '%.1f%%', fontsize = 15,
                   labels = ['Miss', 'Mr', 'Mrs', 'Rare'], subplots = True, ax = axes[0])
axes[0].set_title('Pie Chart Of Title')
axes[0].set_ylabel('')
axes[0].legend()

plot_data = pd.pivot_table(train, index = 'Title', values = 'Survived', aggfunc = 'mean')
sns.barplot(x = plot_data.index, y = 'Survived', data = plot_data, ax = axes[1])
axes[1].set_title('Barplot Of Title By Survived')
axes[1].set_ylabel('')
axes[1].set_xlabel('')

fig.tight_layout()
```

**Fare**

```python
fig, axes = plt.subplots(ncols = 2, figsize = (16, 6))

sns.distplot(train['Fare'], color = 'blue', ax = axes[0])
axes[0].set_title('Distplot of Fare *Train Data*')
axes[0].set_xlabel('')
axes[0].set_ylabel('')

sns.distplot(test['Fare'], color = 'red', ax = axes[1])
axes[1].set_title('Distplot of Fare *Test Data*')
axes[1].set_xlabel('')
axes[1].set_ylabel('')

fig.tight_layout()
```

- **Fare** 변수의 경우 Train, Test 두 개의 데이터에서 왼쪽으로 치우쳐져 있음을 볼 수 있다.

```python
# 'Survived'에 따른 'Fare'의 분포
plt.figure(figsize = (15, 7))

sns.boxplot(x = 'Survived', y = 'Fare', data = train)
plt.title('BoxPlot Of Fare By Survived')
plt.xticks(ticks = [0, 1], labels = ['사망', '생존'])
plt.xlabel('')
plt.ylabel('')
```

- 큰 차이는 보이지 않지만, 생존한 탑승객이 높은 금액을 지불하고 탑승했음을 볼 수 있다.


- Fare 변수가 왼쪽으로 치우쳐져 있으므로, **로그변환**을 통해 치우침을 해결하고 다시 살펴보도록 하자.

```python
# 로그 변환
print('Shape Of Train *Before Create* :', train.shape, '\n')
print('Shape Of Test *Before Create* :', test.shape, '\n')

for dataset in combine:
    dataset['Log_Fare'] = dataset['Fare'] .apply(lambda x: np.log(x) if x > 0 else 0)
    
print('Shape Of Train *After Create* :', train.shape, '\n')
print('Shape Of Test *After Create* :', test.shape)
```

```python
fig, axes = plt.subplots(ncols = 2, figsize = (16, 6))

sns.distplot(train['Log_Fare'], color = 'blue', ax = axes[0])
axes[0].set_title('Distplot of Log_Fare *Train Data*')
axes[0].set_xlabel('')
axes[0].set_ylabel('')

sns.distplot(test['Log_Fare'], color = 'red', ax = axes[1])
axes[1].set_title('Distplot of Log_Fare *Test Data*')
axes[1].set_xlabel('')
axes[1].set_ylabel('')

fig.tight_layout()
```

```python
# 'Survived'에 따른 'Log_Fare'의 분포
# 'Survived'에 따른 'Fare'의 분포
plt.figure(figsize = (15, 7))

sns.boxplot(x = 'Survived', y = 'Log_Fare', data = train)
plt.title('BoxPlot Of Log_Fare By Survived')
plt.xticks(ticks = [0, 1], labels = ['사망', '생존'])
plt.xlabel('')
plt.ylabel('')
```

- 로그변환을 통해 치우쳐짐이 어느정도 해결된 것을 볼 수 있으며, 생존여부에 대해서도 뚜렷한 차이를 볼 수 있다.

```python
# 'Pclass'에 따른 'Log_Fare'의 분포
plt.figure(figsize = (16, 6))

sns.boxenplot(x = 'Pclass', y = 'Log_Fare', data = train)
plt.title('Boxenplot of Log_fare By Pclass')
plt.xticks(ticks = [0, 1, 2], labels = ['Upper', 'Middle', 'Lower'])
plt.xlabel('')
plt.ylabel('')
```

**Embarked**

```python
print('Amount of Null Data *Embarked* :', train['Embarked'].isnull().sum())
```

```python
# 'Embarked'의 분포와 'Embarked'에 따른 생존률 비교
fig, axes = plt.subplots(ncols = 2, figsize = (16, 6))

plot_data = pd.DataFrame(train['Embarked'].value_counts()).sort_index()
plot_data.plot.pie(explode = [0, 0.03, 0.03], autopct = '%.1f%%', shadow = True, subplots = True, fontsize = 15, ax = axes[0])
axes[0].set_title('Pie Chart of Embarked')
axes[0].set_ylabel('')

plot_data = pd.pivot_table(train, index = 'Embarked', values = 'Survived', aggfunc = 'mean')
sns.barplot(x = plot_data.index, y = 'Survived', data = plot_data, ax = axes[1])
axes[1].set_title('Survival Rate By Embarked')
axes[1].set_xlabel('')
axes[1].set_ylabel('')

fig.tight_layout()
```

- Embarked의 **Q와 S**의 경우 생존률이 비슷함을 볼 수 있다.


- S와 Q를 하나의 범주로 묶어주고 다시 살펴보도록 하자.


- 2개의 결측치는 가장 빈도수가 많은 **S**로 채워주도록 하자.

```python
# 결측치 대체
train['Embarked'].fillna('S', inplace = True)

# 결측치 확인
print('Amount of Null Data *Embarked* :', train['Embarked'].isnull().sum())
```

```python
print('Shape Of Train *Before Create* :', train.shape, '\n')
print('Shape Of Test *Before Create* :', test.shape, '\n')

for dataset in combine:
    dataset.loc[(dataset['Embarked'] == 'Q') | (dataset['Embarked'] == 'S'), 'New_embarked'] = 0
    dataset.loc[dataset['Embarked'] == 'C', 'New_embarked'] = 1
    
print('Shape Of Train *After Create* :', train.shape, '\n')
print('Shape Of Test *After Create* :', test.shape)
```

```python
plt.figure(figsize = (10, 6))

plot_data = pd.pivot_table(train, index = 'New_embarked', values = 'Survived', aggfunc = 'mean')

sns.barplot(x = plot_data.index, y = 'Survived', data = plot_data)
plt.title('Survival Rate By New_embarked')
plt.xticks(ticks = [0, 1], labels = ['S & Q', 'C'])
plt.xlabel('')
plt.ylabel('')
```

**Pclass**

```python
fig, axes = plt.subplots(ncols = 2, figsize = (16, 8))

plot_data = train['Pclass'].value_counts().sort_index()
plot_data.plot.pie(explode = [0, 0.03, 0.03], shadow = True, autopct = '%.1f%%', labels = ['Upper', 'Middle', 'Lower'],
                  fontsize = 15, ax = axes[0])
axes[0].set_title('Pie  Chart Of Pclass')
axes[0].set_ylabel('')

plot_data = pd.pivot_table(train, index = 'Pclass', values = 'Survived', aggfunc = 'mean')
sns.barplot(x = plot_data.index, y = 'Survived', data = plot_data, ax = axes[1])
axes[1].set_title('Survival Rate Of Pclass')
axes[1].set_xticklabels(['Upper', 'Middle', 'Lower'])
axes[1].set_xlabel('')
axes[1].set_ylabel('')

fig.tight_layout()
```

- **Lower > Upper > Middle** 순으로 탑승객이 많음을 볼 수 있다.


- Lower Class 에는 탑승객이 많았지만, 생존률은 매우 낮음을 볼 수 있고 **Upper, Middle, Lower** 순으로 생존률이 낮아진다.

```python
# Pclass와 Embarked의 관계
pd.crosstab(train['Pclass'], train['Embarked']).style.background_gradient(cmap = 'Blues', axis = 1)
```

- Pclass에 따라 탑승항구는 차이가 없음을 볼 수 있으며, 일반적으로 **S**에서 많이 탑승함을 볼 수 있다.

- **Pclass**는 Ticket Class를 나타내는데, 연령대가 높으면 Ticket Class가 높아지고, 연령대가 낮으면 Ticket Class 또한 낮아짐을 볼 수 있다.


- **Pclass**와 **Alone**에 따라서 **Age**의 결측값을 채워주도록 하자.

```python
print('*Before* Amount of Null Data *Train Age* :', train['Age'].isnull().sum(), '\n')
print('*Before* Amount of Null Data *Test Age* :', test['Age'].isnull().sum(), '\n')

for alone in train['Alone'].unique():
    for pclass in [1, 2, 3]:
        
        # 'Age' 결측치를 제거한 데이터 추출
        data = train.dropna(subset = ['Age'])
        
        # 'Alone', 'Pclass'에 따른 'Age'의 중앙값 추출
        value = data[(data['Alone'] == alone) & (data['Pclass'] == pclass)]['Age'].median()
        
        # 추출한 값을 Train과 Test 데이터 'Age' 결측값에 대입
        train.loc[(train['Age'].isnull()) & (train['Alone'] == alone) & (train['Pclass'] == pclass), 'Age'] = value
        test.loc[(test['Age'].isnull()) & (test['Alone'] == alone) & (test['Pclass'] == pclass), 'Age'] = value
        
print('*After* Amount of Null Data *Train Age* :', train['Age'].isnull().sum(), '\n')
print('*After* Amount of Null Data *Test Age* :', test['Age'].isnull().sum())
```
